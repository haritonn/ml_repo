## Making some moves w/ scikit-learn

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

X = laptop_df.drop('Price (Euro)', axis = 1)
y = laptop_df['Price (Euro)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=0)
```

### Splitting columns on numerical and categorial features

```python
categorial_cols = [col for col in X.columns if X[col].dtype == 'object']
numerical_cols = [col for col in X.columns if X[col].dtype != 'object']
```

## Using One-Hot-Encoding and StandardScaler to our data 
(so loss of our function should be less)

```python
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown = 'ignore')

preproc = ColumnTransformer(transformers = [('num', num_transformer, numerical_cols),
                                            ('cat', cat_transformer, categorial_cols)])

X_train = preproc.fit_transform(X_train)
X_test = preproc.transform(X_test)
```

### Building Linear Regression

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(X_train, y_train)
preds = linreg.predict(X_test)

data = {'Values': y_test.values, 'Predictions of model': preds}
pd.DataFrame(data)
```

### Values of loss-function could be less via RandomForest model

```python
print('Mean Absolute Error(MAE): ', mean_absolute_error(y_test, preds));
print('R^2 score: ',r2_score(y_test, preds));
print('Mean Squared Error(MSE): ', mean_squared_error(y_test, preds))
```
