user_id contains unique values for each sample (and its corr too low comparing to other features), so it's not good idea to contain them in train/test splits

```python
X = df.drop(['user_id', 'loyalty_score'], axis=1)
y = df['loyalty_score']
```

```python
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
numeric_features = [col for col in X.columns if X[col].dtype != 'object']
categorial_features = [col for col in X.columns if X[col].dtype == 'object']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=17)
```

### Searching for good hyperparameter for Ridge via GridSearchCV

```python
alphas = np.linspace(-3, 3, 200)
ridge_model = Ridge()
searcher = GridSearchCV(ridge_model, [{'alpha': alphas}], cv=10, scoring = "neg_root_mean_squared_error")
temp = StandardScaler()
temp_data = temp.fit_transform(X_train[numeric_features])
searcher.fit(temp_data, y_train)
best_alpha = searcher.best_params_['alpha']

print("Best alpha: %.6f" % best_alpha)
```

Here i decided to use Pipeline

```python
preproc = ColumnTransformer(transformers=[ 
    ('numeric', StandardScaler(), numeric_features),
    ('categorical', OneHotEncoder(), categorical_features)
])

pipeline = Pipeline([
    ('preproc', preproc),
    ('regression', Ridge(alpha=best_alpha))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
data = {'Answers': y_test, 'Predictions': y_pred}
pd.DataFrame(data).reset_index(drop=True).sort_index().head()
```

```python
print("MSE: %.6f" % mean_squared_error(y_pred, y_test));
print("MAE: %.6f" % mean_absolute_error(y_pred, y_test));
print("R2: %.6f" % r2_score(y_pred, y_test))
```

This loss-functions values sounds awesome for us. But what about RidgeCV?

RidgeCV using LeaveOneOut method by default (K = total count of samples)

```python
pipelineCV = Pipeline([
    ('preproc', preproc),
    ('regressCV', RidgeCV())
])

pipelineCV.fit(X_train, y_train)
y_predCV = pipelineCV.predict(X_test)
data = {'Answers': y_test, 'Predictions': y_predCV}
pd.DataFrame(data).reset_index(drop='True').sort_index().head()
```

```python
print("MSE: %.6f" % mean_squared_error(y_predCV, y_test));
print("MAE: %.6f" % mean_absolute_error(y_predCV, y_test));
print("R2: %.6f" % r2_score(y_predCV, y_test))
```
